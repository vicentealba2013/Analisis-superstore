<span style="font-size: 15pt; font-weight: bold; color: #0098cd;">1.3. Generación de variables calculadas.</span>

Genere a partir del dataframe cargado las siguientes variables calculadas.

- Sales/Quantity: Precio de Venta por unidad de producto en cada operación de venta
- Profit/Quantity: Beneficio por unidad de producto en cada operación de venta
- flag_descuento: Marca con "Con Descuento" si esa operacion de venta ha tenido descuento (esto es descuento>0), "Sin Descuento" si no.
- Year: Variable con el año de la fecha de la operación de venta.

Respuesta:

Generamos las variables calculadas que nos pide el ejercicio:

```python
df["Sales/Quantity"] = (df.Sales)/(df.Quantity)
df["Profit/Quantity"] = (df.Profit)/(df.Quantity)
df["Flag_descuento"] = np.where(df["Discount"] > 0, "Con descuento", "Sin descuento")
df["Year"] = df["Order Date"].dt.year.astype(str)

df.head(3)
```

<span style="font-size: 15pt; font-weight: bold; color: #0098cd;">2.1. Gráfico 1.</span>

Respuesta: Vamos a analizar primero si hay relación entre el Precio de Venta por unidad de producto en cada operación de venta medido por la nueva variable que hemos creado llamada Sales/Quantity con la aplicación de descuento en el caso de que así sea.

```python
fig_discount = px.pie(df, names= "Flag_descuento", title = "Ventas con y sin descuento")
fig_discount.show()
```

Conclusiones:


Como podemos ver en el gráfico anterior, la mayoría de las ventas son con descuento (52%). Ello nos indica que dentro de la actividad de ventas de la empresa, las ventas con descuento son muy importantes dentro del volumen total de ventas.

<span style="font-size: 15pt; font-weight: bold; color: #0098cd;">2.2. Gráfico 2.</span>

Respuesta: Vamos a analizar el impacto en el beneficio de la empresa según el tipo de descuento aplicado.

```python
fig_discount_regresion = px.scatter(df, x="Discount", y="Profit", color="Flag_descuento")
fig_discount_regresion.show()
```

Como podemos ver en el gráfico, cuando la empresa no aplica descuentos (descuento = 0), todas las ventas tienen beneficios. Cuando en cambio si se aplican descuentos, vemos que hasta 30% de descuento aproximadamente, la empresa sigue teniendo mayoritariamente beneficios, pero a partir del 40% de descuento, ya registra pérdidas en la mayoría de las ventas. Por lo tanto, podemos concluir que o la empresa no está teniendo en cuenta el margen que tiene realmente al aplicar descuentos grandes o bien, le compensa de forma indirecta por otros medios.

Vamos a ver ahora el beneficio anual en ventas con y sin descuento

```python
df_agrupado = df.groupby(["Year", "Flag_descuento"])["Profit"].mean().reset_index()
df_agrupado.head(10)

df_agrupado = px.bar(
    df_agrupado,
    x="Year",
    y="Profit",
    color="Flag_descuento",
    barmode="group",
    title="Beneficio Promedio por año (Con/Sin descuento)",
    labels ={"Profit": "Beneficio Promedio", "Year": "Año"}
)
df_agrupado.show()
```

Conclusiones:


Como podemos ver, el beneficio medio de las ventas sin descuento es positivo mientras que las ventas con descuento registran márgenes medios negativos. Si bien entre 2014 y 2016 se reducen las pérdidas medias en las ventas con descuento, en 2017 vuelven a repuntar ligeramente. Las ventas sin descuento también mejoran su margen medio entre 2014 y 2016 pero en 2017 bajan un poco.

<span style="font-size: 15pt; font-weight: bold; color: #0098cd;">2.3. Gráfico 3.</span>

Respuesta: Vamos a analizar la correlación entre ventas y beneficio por categorias con y sin descuento

```python
correlacion = df["Sales/Quantity"].corr(df["Profit/Quantity"])
print(f"El coeficiente de correlación es: {correlacion}")

fig =px.scatter (df,
                x="Sales/Quantity",
                y="Profit/Quantity",
                color="Flag_descuento",
                symbol="Category",
                trendline="ols",
                title="Relación entre Descuento y Beneficio")
fig.update_layout(width=900, height=500)
fig.show()
```

Viendo la correlación por categorías entre beneficio y ventas, con y sin descuento aplicado, podemos observar que en general, las categorías sin descuento tienen una correlación positiva, con lo que interesa seguir sin aplicar descuentos, ya que la tendencia es claramente al alza. En cuanto a la correlación en las categorías con descuento, vemos claramente una tendencia cercana a 0, por lo que no podríamos asegurar que a mayor descuento, mayores ventas.

Conclusiones:

Viendo toda la información que nos detallan los gráficos anteriores, podemos concluir que:

- La empresa tiene un mayor volumen de ventas en artículos con descuento (52%)
- A partir del 30% de descuento, las ventas en las que se aplique registran mayoritariamente pérdidas.
- Analizando el beneficio promedio anual, vemos que entre 2014 y 2016 se redujeron las pérdidas medias en las ventas con descuento y se mejoró el beneficio medio de las ventas sin descuento, aunque en 2017 hubo una ligera reducción de márgenes en ambos.
- Podemos concluir que no hay una correlación positiva entre descuento y beneficios, por lo que a mayores descuentos no conseguiremos mejores resultados.
